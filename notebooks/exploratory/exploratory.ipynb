```python
# Importing libraries and Setting Dislplay Options

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from bs4 import BeautifulSoup
import requests

#Setting Options
pd.set_option ('display.max_columns', 75)
pd.set_option('display.max_rows', 100)
```

```python
# reading data
df = pd.read_csv('processed_data.csv')
df.head()
```

# Exploratory Data Analysis

```python
df.info()
```

```python
#Descriptive Statistics of numeric columns
df.describe()
```

```python
#Top 5 Valuable Player
df.nlargest(5,columns="Value")[["Name","Age","Team","Value"]]
```

```python
#Top 5 Rating Player
df.nlargest(5,columns="Overall")[["Name","Age","Team","Overall"]]
```

```python
#Top 5 Clup (Total Value)

pd.DataFrame(df.groupby("Team").Value.sum().sort_values(ascending=False).head())
```

```python
#Age , Height and Weight Stats
df[["Age","Height","Weight"]].describe()
```

## Visualization

```python
# create a correlation matrix
corr_mat = df.corr()
corr_mat
```

```python
#Correlation With Value (most correlated with positive)
pd.DataFrame(corr_mat["Value"]).sort_values("Value", ascending=False).head(10)
```

```python
#Correlation With Value (modes correlated with negative)
pd.DataFrame(corr_mat["Value"]).sort_values("Value", ascending=True).head(7)
```
